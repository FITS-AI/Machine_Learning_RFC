```python
import numpy as np
import pandas as pd
```

```python
nutcomp = "/content/drive/MyDrive/nutcomp.csv"
nutcomp_df = pd.read_csv(nutcomp)
nutcomp_df.head()
```

```python
# Pencarian Duplikasi Data
print("Jumlah duplikasi: ", nutcomp_df.duplicated().sum())
```

```python
# Menemukan Missing Value
nutcomp_df.isnull().sum()
```

```python
# Mengisi nilai kosong pada kolom 'Sugar_Tot_In_Grams' & 'Fiber_TD_In_Grams' dengan 0
nutcomp_df['Sugar_Tot_In_Grams'] = nutcomp_df['Sugar_Tot_In_Grams'].fillna(0)
nutcomp_df['Fiber_TD_In_Grams'] = nutcomp_df['Fiber_TD_In_Grams'].fillna(0)

# Menampilkan DataFrame
print(nutcomp_df)
```

```python
# Menghapus kolom dengan nilai kosong
nutcomp_df_cleaned = nutcomp_df.dropna(axis=1)

# Menampilkan DataFrame yang sudah dibersihkan
nutcomp_df_cleaned.info()
```

```python
# Menampilkan DataFrame yang sudah dibersihkan
nutcomp_df_cleaned.head()
```

```python
# Mengubah nama kolom pada DataFrame
nutcomp_df_cleaned = nutcomp_df_cleaned.rename(columns={
    'NDB_No': 'food_id',
    'Short_Description': 'Composition',
    'Energy_In_Kilogram_Calorie': 'Calories',
    'Protein_In_Grams': 'Protein',
    'Lipid_Tot_In_Grams': 'Fat',
    'Carbohydrate_In_Grams': 'Carbohydrate',
    'Sugar_Tot_In_Grams': 'Sugar',
    'Fiber_TD_In_Grams': 'Fiber'
})

# Menampilkan DataFrame dengan nama kolom baru
nutcomp_df_cleaned.head()
```
